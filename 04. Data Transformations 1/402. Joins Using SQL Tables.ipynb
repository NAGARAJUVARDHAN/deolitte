
#### Types of joins in Spark

- inner join (default) 
- cross join
- outer/ full/ fullouter/ full_outer
- left/ leftouter/ left_outer
- right/ rightouter/ right_outer
- semi/ leftsemi/ left_semi
- anti/ leftanti/ left_anti

```sql
select products.*,brands.brand_name from products inner join brands on products.brand_id = brands.brand_id;
```

```sql
select products.*,brands.brand_name from products join brands on products.brand_id = brands.brand_id;
```

```sql
select products.*,brands.brand_name from products full outer join brands on products.brand_id = brands.brand_id;
```

```sql
select products.*,brands.brand_name from products left outer join brands on products.brand_id = brands.brand_id;
```

```sql
select products.*,brands.brand_name from products right outer join brands on products.brand_id = brands.brand_id;
```

```sql
select products.* from products left semi join brands on products.brand_id = brands.brand_id;
```

```sql
select products.* from products left anti join brands on products.brand_id = brands.brand_id;
```
