```python
from pyspark.sql import functions as F
```

```python
emp_df.select(F.count('*')).show()   
```

```python
emp_df.groupBy('deptno').agg(F.count('*')).show()  
```

```python
# get salary summation for a given deptno

emp_df.filter("deptno = 10").select(F.sum('sal').alias ('total_sal')).show()
```

```python
emp_df.groupBy("deptno").agg(F.sum('sal').alias ('avg_sal')).show()
```

```python
# get count of employees, total_sal, min_sal, max_sal avg_sal for dept 10
(emp_df.filter('deptno=10')
 .select(F.count('sal').alias('emp_count'),
         F.sum('sal').alias('total_sal'),
         F.min('sal').alias('min_sal'),
         F.max('sal').alias('max_sal'),
         F.avg('sal').alias('avg_sal'),
         F.stddev('sal').alias('std_sal'),
         F.variance('sal').alias('var_sal')
        )
 ).show()
```

```python
# get department wise summary
# get count of employees, total_sal, min_sal, max_sal avg_sal for dept 10
(emp_df.groupBy('deptno')
 .agg(F.count('sal').alias('emp_count'),
         F.sum('sal').alias('total_sal'),
         F.min('sal').alias('min_sal'),
         F.max('sal').alias('max_sal'),
         F.avg('sal').alias('avg_sal'),
         F.stddev('sal').alias('std_sal'),
         F.variance('sal').alias('var_sal')
        )
 ).show()
```
