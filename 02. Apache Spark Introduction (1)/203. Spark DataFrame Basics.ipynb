```python
# data with date and time types

import datetime
users_data = [(1, 'Ajit', 'aaa@xyz.com', True, 1002.55, datetime.date(2021, 2, 10), datetime.datetime(2021, 4, 11, 12, 30,  40)),
        (2, 'Brijesh', 'bbb@xyz.com', False, 11122.55, datetime.date(2021, 3, 11), datetime.datetime(2021, 5, 16, 12, 30,  40))]

user_schema = '''
id INT,
f_name STRING,
email STRING,
is_customer BOOLEAN,
amount_paid FLOAT,
customer_from DATE,
last_updated_ts TIMESTAMP
'''
```

```python
df2 = spark.createDataFrame(users_data, schema=user_schema)
df2.show()
```

```python
from pyspark.sql.types import *

fields =StructType([
    StructField('id', IntegerType()),
     StructField('f_name', StringType()),
     StructField('email', StringType()),
     StructField('is_customer', BooleanType()),
     StructField('amount_paid', FloatType()),
     StructField('customer_from', DateType()),
     StructField('last_updated_ts', TimestampType()),
])

type(fields)
```

```python
df3 = spark.createDataFrame(users_data, schema=fields)
df3.show()
```

Type you mean
Spark SQL type
What it represents
How you work with it




string
StringType
One atomic text value per row (e.g., "555-1212")
Use string functions (e.g., trim/regex/substring); no unnesting


list (in DF)
ArrayType(<elementType>)
An ordered collection of 0..N values per row (all same element type), e.g., ["555-1","555-2"]
Use array functions (size, indexing, transform, filter) or explode to turn elements into rows


struct
StructType([field1, field2, ...])
A nested record/object per row with named fields (like a JSON object), e.g., {type:"mobile", number:"555-1"}
Access fields by name (col("s.field")); you can explode an array of structs
