```python
from pyspark.sql.functions import explode
# explode () is a function that unnests an array or map column into multiple rows—one row per array element (or map entry), while repeating the other columns.
## will not make changes in-place
## withColumn is used to apply any transformations on existing Dataframe column and create new column out of it
## syntax: withColumn(column_name, any_operation_on_column)

df.withColumn('phone', explode('phone_numbers')).drop('phone_numbers').show()
```

### Map Type

```python
# pay attention to 'phone_numbers' field


import datetime
user_list = [
    {
        'id': 1,
        'f_name': 'John',
        'l_name': 'Maddocks',
        'email': 'abc@xyz.com',
        'phone_numbers': {'mobile':'+1 234 567 8901', 'home':'+1 234 444 5556'},
        'is_customer': True,
        'amount_paid': 1001.22,
        'customer_from': datetime.date(2020, 2, 1),
        'last_updated': datetime.datetime(2021, 3, 4, 10, 40, 25)
    },
    {
        'id': 2,
        'f_name': 'Bob',
        'l_name': 'Miller',
        'email': 'abc@xyz.com',
        'phone_numbers': {'mobile':'+1 234 567 9999', 'home':'+1 234 444 7777'},
        'is_customer': False,
        'amount_paid': 1000.5,
        'customer_from': datetime.date(2021, 6, 1),
        'last_updated': datetime.datetime(2022, 3, 4, 10, 40, 25)
    },
        {
        'id': 3,
        'f_name': 'Chris',
        'l_name': 'Miller',
        'email': 'abc@xyz.com',
        'phone_numbers': {'mobile':'+1 234 444 7777'},
        'is_customer': True,
        'amount_paid': 1004.5,
        'customer_from': datetime.date(2021, 6, 1),
        'last_updated': datetime.datetime(2022, 3, 4, 10, 40, 25)
    },
        {
        'id': 4,
        'f_name': 'christy',
        'l_name': 'Marsh',
        'email': 'abc@xyz.com',
        'phone_numbers': None,
        'is_customer': True,
        'amount_paid': 1004.5,
        'customer_from': datetime.date(2021, 6, 1),
        'last_updated': datetime.datetime(2022, 3, 4, 10, 40, 25)
    }
]
```

```python
df = spark.createDataFrame(user_list)
df.show()
```

```python
df.dtypes
```

```python
df.select('id', 'phone_numbers').show(truncate=False)
```

```python
from pyspark.sql.functions import col

df.select(col('phone_numbers')['mobile']).show()
```

```python
from pyspark.sql.functions import col

df.select(col('phone_numbers')['home']).show()
```

```python
df.select('id', col('phone_numbers')['mobile'].alias('mobile'), col('phone_numbers')['home'].alias('home')).show()
```

```python
# it is used to give alternate name to table or column

df.alias('u').select('u.id', 'u.f_name', 'u.l_name').show()
```

```python
# it create two columns from 'phone_numbers'--> key and value
from pyspark.sql.functions import explode
df.select('id', explode('phone_numbers')). \
drop('phone_numbers'). \
show()
```

```python
# it create two columns from 'phone_numbers'--> key and value
from pyspark.sql.functions import explode

(df.select('id', explode('phone_numbers'))
   .drop('phone_numbers')
   .show()
)
```

```python
df1=df.select('id', explode('phone_numbers'))
df1.show()
```

```python
df1.withColumnRenamed('key', 'phone_type'). \
withColumnRenamed('value', 'phone_number'). \
show()
```

```python
# withColumnRenamed is used to rename existing column of dataframe
# it doesn't rename the column inplace, it creates another dataframe
# syntax : withColumnRenamed(old_Col_Name, new_Col_Name)

df.select('id', explode('phone_numbers')). \
withColumnRenamed('key', 'phone_type'). \
withColumnRenamed('value', 'phone_number'). \
show()
```

```python
df.select('*', explode('phone_numbers')). \
withColumnRenamed('key', 'phone_type'). \
withColumnRenamed('value', 'phone_number'). \
drop('phone_numbers'). \
show()
```

```python
#explode(col) drops rows where col is null or empty.
#explode_outer(col) preserves them (returns a single row with null).
from pyspark.sql.functions import explode_outer

df.select('*', explode_outer('phone_numbers')). \
withColumnRenamed('key', 'phone_type'). \
withColumnRenamed('value', 'phone_number'). \
drop('phone_numbers'). \
show()
```


### Struct Type

```python
import datetime
from pyspark import Row

user_list = [
    {
        'id': 1,
        'f_name': 'John',
        'l_name': 'Maddocks',
        'email': 'abc@xyz.com',
        'phone_numbers': Row(mobile='+1 234 567 8901', home='+1 234 444 5556'),
        'is_customer': True,
        'amount_paid': 1001.22,
        'customer_from': datetime.date(2020, 2, 1),
        'last_updated': datetime.datetime(2021, 3, 4, 10, 40, 25)
    },
    {
        'id': 2,
        'f_name': 'Bob',
        'l_name': 'Miller',
        'email': 'abc@xyz.com',
        'phone_numbers': Row(mobile='+1 234 567 9999', home='+1 234 444 7777'),
        'is_customer': False,
        'amount_paid': 1000.5,
        'customer_from': datetime.date(2021, 6, 1),
        'last_updated': datetime.datetime(2022, 3, 4, 10, 40, 25)
    },
        {
        'id': 3,
        'f_name': 'Chris',
        'l_name': 'Miller',
        'email': 'abc@xyz.com',
        'phone_numbers': Row(mobile='+1 234 444 7777', home=None),
        'is_customer': True,
        'amount_paid': 1004.5,
        'customer_from': datetime.date(2021, 6, 1),
        'last_updated': datetime.datetime(2022, 3, 4, 10, 40, 25)
    },
        {
        'id': 4,
        'f_name': 'christy',
        'l_name': 'Marsh',
        'email': 'abc@xyz.com',
        'phone_numbers': Row(mobile=None, home=None),
        'is_customer': True,
        'amount_paid': 1004.5,
        'customer_from': datetime.date(2021, 6, 1),
        'last_updated': datetime.datetime(2022, 3, 4, 10, 40, 25)
    }
]
```

```python
df = spark.createDataFrame(user_list)
```

```python
# phone_numbers is struct type

df.dtypes
```

```python
df.select('id', 'phone_numbers').show(truncate=False)
```

```python
df.select('id', 'phone_numbers.mobile', 'phone_numbers.home').show()
```

```python
from pyspark.sql.functions import col

df.select('id', col('phone_numbers')['mobile'], col('phone_numbers')['home']).show()
```

```python
df.select('id', col('phone_numbers.*')).show()
```

```python
df.select('id', explode('phone_numbers')).show()
```

```python
help(explode)
```
