```python
#rdd = sc.parallelize(range(1, 10))
from pyspark.sql import SparkSession
rdd = spark.createDataFrame(range(1,20))
```

Now we will build RDD pipeline

```python
def odd(n):
    print('odd called', n)
    return n%2 == 1

rdd2 = rdd.filter(odd)
```

```python
def mul(n):
    print('mul called', n)
    return n*10

rdd3 = rdd2.map(mul)
```

```python
## Without cache
# it will create 3 independent jobs

print('count', rdd3.count())
print('sum', rdd3.sum())
print('min', rdd3.min())
```

```python
print('is cached', rdd3.is_cached)
```

```python
## Cache the results
rdd3.cache()
```

```python
from pyspark import StorageLevel

#rdd = sc.parallelize([1, 2, 3, 4])
rdd.persist(StorageLevel.MEMORY_AND_DISK)
```

### cache() Vs. persist()
- RDD.cache() caches the RDD with the default storage level MEMORY_ONLY
- DataFrame.cache() caches the DataFrame with the default storage level MEMORY_AND_DISK
- The persist() method is used to store it to the user-defined storage level

```python
rdd3.cache()
print(rdd3.getStorageLevel())
```

```python
from pyspark import StorageLevel
rdd2.persist(StorageLevel.MEMORY_AND_DISK)
print(rdd2.getStorageLevel())
```

```python
rdd2.unpersist()
rdd2.persist(StorageLevel.MEMORY_AND_DISK_2)
print(rdd2.getStorageLevel())
```
