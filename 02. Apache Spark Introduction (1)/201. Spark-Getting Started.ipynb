
##Role of Partitions in Spark 

**1 partition = 1 task**

### How to increase/decrease the partitions at runtime


```python
rdd = sc.parallelize(range(1, 1000), 1)

# all numbers in 1 partition
# 8 parallel tasks at a time
# only one task = due to 1 partitions
# task completion takes time
# poor resource planning, other parallel threads do not have work

print("Rdd: ", rdd.getNumPartitions())
```

```python
# increase the partitions to 8

rdd2 = rdd.repartition(8)

print(rdd2.getNumPartitions())
```

```python
rdd = sc.parallelize(range(1, 5), 20)
rdd.glom().collect()
```

```python
rdd2 = rdd.repartition(4)
rdd2.getNumPartitions()
```

```python
rdd2 = rdd.coalesce(2)
rdd2.getNumPartitions()
```

```python
rdd2.glom().collect()
```

```python
rdd3 = rdd2.coalesce(10)
rdd3.getNumPartitions()
```

```python
rdd3.glom().collect()
```


### Repartition vs. Coalesce

1. Repartition results into full shuffle and coalesce does not
2. Coalecse is efficient as compaired to repartition
3. repartition guarentees balanced partitions and coalesce might generrate imbalanced partitions
4. Repartition is suitable for increasing as well as descresing the numbe rof partitions
5. Use repartition when changing the number of partitions or evenly distributing data across partitions
6. Use coalesce when decreasing the number of partitions without incurring the cost of a full shuffle
