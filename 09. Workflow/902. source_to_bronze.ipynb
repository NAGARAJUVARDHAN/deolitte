```python
# giving a path till folder only (it will read all the files in that folder)

source_path_movies = "/Volumes/workspace/default/myspace/movies/movies 1.csv"
source_path_ratings = "/Volumes/workspace/default/myspace/movies/ratings 1.csv"

moviesDF = (spark.read.format("csv")
                 .option("header", "true")
                 .option("inferSchema", "true")
                 .load(source_path_movies)
                 )

ratingsDF = (spark.read.format("csv")
                 .option("header", "true")
                 .option("inferSchema", "true")
                 .load(source_path_ratings)
                 )
```

```python
moviesDF.show(5)
```

```python
ratingsDF.show(5)
```

```python
bronze_path_movies = "/Volumes/workspace/default/myspace/movies/bronze/movies/movies.parquet"
bronze_path_ratings = "/Volumes/workspace/default/myspace/movies/bronze/ratings/ratings.parquet"

moviesDF.write.parquet(path=bronze_path_movies,
                        mode='overwrite')

ratingsDF.write.parquet(path=bronze_path_ratings,
                        mode='overwrite')
```
