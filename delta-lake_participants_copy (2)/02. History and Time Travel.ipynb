## History and Time Travel
1. Query Delta Lake table history
1. Query previous versions of the data
1. Query data from a specific time. 
1. Restore data to a specific version.

#### 1. Query Delta Lake Table History

```sql
DESCRIBE HISTORY p101.p101schema.companies;
```

#### 2. Query Data from a Specific Version

```sql
SELECT * FROM p101.p101schema.companies;
```

```sql
select * from p101.p101schema.companies 
version as of 1;
```

#### 3. Query Data from a Specific Time

```sql
SELECT * FROM p101.p101schema.companies
TIMESTAMP AS OF '2026-04-08T06:45:51.000+00:00';  
-- upto this time how many records will be inserted or any other things it see only last entry time 
```

#### 4. Restore Data in the Table to a Specific Version

```sql
RESTORE TABLE p101.p101schema.companies VERSION AS OF 2;
```
