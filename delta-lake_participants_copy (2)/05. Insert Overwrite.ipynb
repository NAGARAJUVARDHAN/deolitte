## Insert Overwrite
1. Replace all the data in a table
1. Replace all the data from a specific partition
1. How to handle schema changes

INSERT OVERWITE - Overwrites the existing data in a table or a specific partition with the new data. 

INSERT INTO - Appends new data


### 1. Replace all the data in a table

```sql
DROP TABLE IF EXISTS p101.p101schema.gold_companies;

CREATE TABLE p101.p101schema.gold_companies
  (company_name STRING,
   founded_date DATE,
   country      STRING);

INSERT INTO p101.p101schema.gold_companies 
(company_name, founded_date, country)
VALUES ("Apple", "1976-04-01", "USA"),  
       ("Tencent", "1998-11-11", "China"); 

SELECT * FROM p101.p101schema.gold_companies;        
```

```sql
DROP TABLE IF EXISTS p101.p101schema.bronze_companies;

CREATE TABLE p101.p101schema.bronze_companies
  (company_name STRING,
   founded_date DATE,
   country      STRING);

INSERT INTO p101.p101schema.bronze_companies 
(company_name, founded_date, country)
VALUES ("Apple", "1976-04-01", "USA"),
       ("Microsoft", "1975-04-04", "USA"),
       ("Google", "1998-09-04", "USA"),
       ("Amazon", "1994-07-05", "USA"),
       ("Tencent", "1998-11-11", "China");   

SELECT * FROM p101.p101schema.bronze_companies;       
```

```sql
insert into table p101.p101schema.gold_companies
select * from p101.p101schema.bronze_companies;
```

```sql
DROP TABLE IF EXISTS p101.p101schema.gold_companies;

CREATE TABLE p101.p101schema.gold_companies
  (company_name STRING,
   founded_date DATE,
   country      STRING);

INSERT INTO p101.p101schema.gold_companies 
(company_name, founded_date, country)
VALUES ("Apple", "1976-04-01", "USA"),  
       ("Tencent", "1998-11-11", "China"); 

SELECT * FROM p101.p101schema.gold_companies;   
```

```sql
insert overwrite table p101.p101schema.gold_companies
select * from p101.p101schema.bronze_companies;
```

```sql
select * from p101.p101schema.gold_companies;
```

### 2. Replace all the data from a specific partition

```sql
DROP TABLE IF EXISTS p101.p101schema.gold_companies_partitioned;

CREATE TABLE p101.p101schema.gold_companies_partitioned
  (company_name STRING,
   founded_date DATE,
   country      STRING)
PARTITIONED BY (country);

INSERT INTO p101.p101schema.gold_companies_partitioned 
(company_name, founded_date, country)
VALUES ("Apple", "1976-04-01", "USA"),  
       ("Tencent", "1998-11-11", "China"); 

SELECT * FROM p101.p101schema.gold_companies_partitioned;        
```

```sql
DESC EXTENDED p101.p101schema.gold_companies_partitioned
```

```sql
DROP TABLE IF EXISTS p101.p101schema.bronze_companies_usa;

CREATE TABLE p101.p101schema.bronze_companies_usa
  (company_name STRING,
   founded_date DATE,
   country      STRING);

INSERT INTO p101.p101schema.bronze_companies_usa 
(company_name, founded_date, country)
VALUES ("Apple", "1976-04-01", "USA"),
       ("Microsoft", "1975-04-04", "USA"),
       ("Google", "1998-09-04", "USA"),
       ("Amazon", "1994-07-05", "USA");   

SELECT * FROM p101.p101schema.bronze_companies_usa;       
```

```sql
insert overwrite table p101.p101schema.gold_companies_partitioned
partition (country = 'USA') 
select company_name, founded_date
from p101.p101schema.bronze_companies_usa;
```

```sql
SELECT * FROM p101.p101schema.gold_companies_partitioned;
```

### 3. How to handle schema changes

Insert Overwrite -> Use to overwrite the data in a table or a partition when there are no schema changes. 

Insert overwrite does not support if schemas of both tables are different 

Create or replace table -> Use when there are schema changes.
